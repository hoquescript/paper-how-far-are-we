```python
import pandas as pd
import numpy as np
```

```python
java_codes = pd.read_json("../data/java/test.jsonl", lines=True)
python_codes = pd.read_json("../data/python/test.jsonl", lines=True)
```

```python
print(java_codes.shape)
print(java_codes.columns)
print(java_codes.iloc[0])
```

```python
print(python_codes.shape)
print(python_codes.columns)
print(python_codes.iloc[0])
```

```python
java_slice = java_codes.sample(25)
java_slice['language'] = "java"
python_slice = python_codes.sample(25)
python_slice['language'] = "python"

merged_slice = pd.concat([java_slice,python_slice])
```

```python
merged_slice.to_json("../data/sample_data.jsonl", orient="records", lines=True)
```
